```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

numerical = pd.read_csv('datalab2_numerical.csv')
numerical_x=pd.read_csv('datalab2_numerical_x.csv')

categorical = pd.read_csv('datalab2_categorical.csv')
categorical = categorical.drop(['Unnamed: 0','customer','effective to date'],axis=1)
categorical
```

```python
one_hot_names = []
for col in categorical.columns:
    col_uniques = sorted(categorical[col].unique())
    for unique in col_uniques:
        one_hot_names.append(unique)

one_hot_names
```

```python
categorical_x = pd.DataFrame(OneHotEncoder().fit(pd.DataFrame(categorical))
                             .transform(pd.DataFrame(categorical)).toarray())
categorical_x.columns = one_hot_names
```

```python
categorical_x
```

```python
x = pd.DataFrame(np.concatenate((numerical_x,categorical_x),axis=1))
y = numerical['total claim amount']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=100)
```

```python
print(len(x))
print(len(x_train))
print(len(x_test))
```

```python
import statsmodels.api as sm

model = sm.OLS(y_train,x_train).fit()
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
y
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
import math
```

```python
r2 = r2_score(y_test,predictions)
r2
```

```python
mse = mean_squared_error(y_test,predictions)
mse
```

```python
rmse = math.sqrt(mse)
rmse
```
